# 分析1：將三組不同的資料檔案，合併在同一DataFrame，df4
檔案：https://acupun.site/lecture/pandas/example/resource/students.xls

資料表：105級

106級

107級

```python
import pandas as pd
df1=pd.read_excel('https://acupun.site/lecture/pandas/example/resource/students.xls','105級')
df2=pd.read_excel('https://acupun.site/lecture/pandas/example/resource/students.xls','106級')
df3=pd.read_excel('https://acupun.site/lecture/pandas/example/resource/students.xls','107級')
df4=pd.concat([df1,df2,df3],axis=0,ignore_index=True)
#print(df4.isnull().sum().sum())
df4
```

# 分析2：入學年份的人數統計

```python
df4.groupby('入學年份')['名字'].size()\
  .rename(index="人數",inplace=True)\
  .to_frame()
```

# 分析4：畫圖：入學年份的人數統計

```python
import matplotlib.pyplot as plt
df4.groupby("入學年份")["名字"].size()\
  .reset_index()\
  .rename(columns={"名字":"人數"})\
  .plot(kind="bar",legend=True,rot=0,x="入學年份",y="人數",title='圖4：入學年份的人數統計')
plt.xlabel("入學年份")
plt.ylabel("人數")
plt.show()
```

# 分析5：分析『不同年份，不同男女』人數，做樞紐分析表

```python
df5 = df4.groupby(["性別","入學年份"])["名字"].size()\
  .unstack(level="入學年份")
df5.columns=[["人數","人數","人數"],["105年","106年","107年"]]
df5
```

# 分析6：分析『不同年份，不同男女』人數，做交叉分析圖

```python
import matplotlib.pyplot as plt
df5=df4.groupby(["入學年份","性別"]).agg({"名字":"size"})\
  .unstack(level="入學年份")
df5.columns=['105年','106年','107年']
df5.plot(kind="bar",legend=True,rot=0,title='圖6：分析『不同年份，不同男女』人數')
plt.xlabel("性別")
plt.ylabel("人數")
plt.show()
```

# 分析11：用groupby來分析不同『入學年份，入學管道』人數(0位小數點)，做樞紐分析表，缺值填入0

```python
df6=df4.groupby(['入學管道','入學年份'])["名字"].size()\
  .unstack(level="入學年份")
df6.fillna(0,inplace=True)
df6.columns=[["","","人數"],["105年","106年","107年"]]
df6=df6.astype(int)
df6

# df5 = df4.groupby(["性別","入學年份"])["名字"].size()\
#   .unstack(level="入學年份")
# df5.columns=[["人數","人數","人數"],["105年","106年","107年"]]
# df5
```

# 分析12：用pivot_table指令來分析不同『入學年份，入學管道』人數(0位小數點)，做樞紐分析表，缺值填入0

```python
df7 = df4.pivot_table(columns="入學年份", index="入學管道", values="名字", aggfunc="size")
df7.fillna(0,inplace=True)
df7.columns =["105","106","107"]
df7.columns.name=('入學年份')
df7 = df7.astype(int)
df7
```

# 分析14：請讀入另外一個資料表：students.xls, 學校地區表，df5

# 分析15：請合併前面的df4,df5，結合成df6（合併方式以df4為準)
