#10.讀入企業的2種國外ESG評比分數，然後自訂一個ESG綜合指標，把2種不同的ESG分數合併計算，根據綜合ESG來排序(高到低)，標註顏色

（1）把MSCI ESG換算分數：

AAA = 100分

AA = 90分

A = 80分

B = 40分

（2）綜合ESG分數 = （MSCI+標普全球)/2

檔案：https://acupun.site/lecture/pandas/midterm/2023/midterm.xlsx

資料表：ESG

```python
import pandas as pd
df=pd.read_excel('https://acupun.site/lecture/pandas/midterm/2023/midterm.xlsx','ESG')
df


def grade(x):
  if x=='AAA':
    return 100
  elif x=='AA':
    return 90
  elif x=='A':
    return 80
  elif x=='B':
    return 40
  else:
    return ''

df['MSCI分數']=df['MSCI'].map(grade)
df

df['綜合ESG分數']=(df['MSCI分數']+df['標普全球'])/2
df.sort_values('綜合ESG分數',ascending=False).style.highlight_between(subset='綜合ESG分數',color='yellow')
```

#16.顯示『亞洲』國家資料，根據AQI指數排序，根據AQI指數顯示不同的背景顏色
若是AQI指數有缺值，則刪除該筆

檔案：https://acupun.site/lecture/pandas/homework/hw3/airPollution.xlsx

資料表：空氣污染指數

咖啡色：AQI指數，201～300

紅色：AQI指數，151～200

橘色：AQI指數，101～150

黃色：AQI指數，51～100

淺綠色：AQI指數，0～50

```python
import pandas as pd
df=pd.read_excel('https://acupun.site/lecture/pandas/homework/hw3/airPollution.xlsx','空氣污染指數')
df

df.dropna(subset='AQI污染指數', axis='index', inplace=True)
df

df2=df[df['區域']=='亞洲']
df3=df2.sort_values('AQI污染指數')
df3

def mycolor(x):
  if 300>=x>=201:
    return 'background-color:brown'
  elif 200>=x>=151:
    return 'background-color:red'
  elif 150>=x>=101:
    return 'background-color:orange'
  elif 100>=x>=51:
    return 'background-color:yellow'
  elif 50>=x>=0:
    return 'background-color:lightgreen'
  else:
    return ''

df3.style.applymap(mycolor,subset='AQI污染指數')
```
